```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

digits = load_digits()
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.3, random_state=42)
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print('Accuracy: ', accuracy)
print('Precision: ', precision)
print('Recall: ', recall)
print('F1 Score: ', f1)
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
import seaborn as sns

digits = load_digits()

X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.3, random_state=42)

dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train, y_train)
y_pred_dt = dt_clf.predict(X_test)

print("Confusion matrix for Decision Tree Classifier:")
print(confusion_matrix(y_test, y_pred_dt))

print("\nFeature importance (aka Gini importance) for each feature in Decision Tree Classifier:")
for feature, importance in zip(digits.feature_names, dt_clf.feature_importances_):
    print(feature, ":", importance)
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV

# Load the dataset
digits = load_digits()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)

# Tune the Decision Tree Classifier using Grid Search
dtc_params = {'criterion': ['gini', 'entropy'], 'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10]}
dtc = DecisionTreeClassifier(random_state=42)
dtc_gs = GridSearchCV(dtc, dtc_params, cv=5)
dtc_gs.fit(X_train, y_train)

# Print the best hyperparameters
print('Best Decision Tree Classifier hyperparameters:', dtc_gs.best_params_)

# Train and evaluate the Decision Tree Classifier on the test set
dtc_best = DecisionTreeClassifier(**dtc_gs.best_params_, random_state=42)
dtc_best.fit(X_train, y_train)
dtc_y_pred = dtc_best.predict(X_test)
dtc_cm = confusion_matrix(y_test, dtc_y_pred)
print('Decision Tree Classifier Confusion Matrix:')
print(dtc_cm)

print('\nFeature importance of each feature in the dataset:')
for feature_name, importance in zip(digits.feature_names, dtc_best.feature_importances_):
    print(f'{feature_name}: {importance:.3f}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Plot confusion matrix for decision tree classifier
plt.figure(figsize=(5,4))
sns.heatmap(dtc_cm, annot=True, cmap='Blues', fmt='g')
plt.title('Decision Tree Classifier Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

housing = fetch_california_housing()
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(housing.data, housing.target, test_size=0.2, random_state=42)

# Decision Tree Regressor
dt_params = {'max_depth': [3, 5, 7, 10]}
dt_grid = GridSearchCV(DecisionTreeRegressor(), dt_params, cv=5, scoring='neg_mean_squared_error')
dt_grid.fit(X_train, y_train)

dt_reg = dt_grid.best_estimator_
dt_pred = dt_reg.predict(X_test)

print("Decision Tree Regressor:")
print("Best Hyperparameters: ", dt_grid.best_params_)
print("MSE: ", mean_squared_error(y_test, dt_pred))
print("R^2: ", r2_score(y_test, dt_pred))

# Feature Importance for Decision Tree Regressor
importances_dt = dt_reg.feature_importances_
features_dt = housing.feature_names
indices_dt = np.argsort(importances_dt)[::-1]

print("\nFeature Importances (Decision Tree Regressor):")
for i in range(len(features_dt)):
    print("%s : %f" % (features_dt[indices_dt[i]], importances_dt[indices_dt[i]]))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

housing = fetch_california_housing()
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(housing.data, housing.target, test_size=0.2, random_state=42)

rf_params = {'n_estimators': [50, 100, 150], 'max_depth': [3, 5, 7, 10]}
rf_grid = GridSearchCV(RandomForestRegressor(), rf_params, cv=5, scoring='neg_mean_squared_error')
rf_grid.fit(X_train, y_train)

rf_reg = rf_grid.best_estimator_
rf_pred = rf_reg.predict(X_test)

print("Random Forest Regressor:")
print("Best Hyperparameters: ", rf_grid.best_params_)
print("MSE: ", mean_squared_error(y_test, rf_pred))
print("R^2: ", r2_score(y_test, rf_pred))

# Feature Importance for Random Forest Regressor
importances_rf = rf_reg.feature_importances_
features_rf = housing.feature_names
indices_rf = np.argsort(importances_rf)[::-1]

print("\nFeature Importances (Random Forest Regressor):")
for i in range(len(features_rf)):
    print("%s : %f" % (features_rf[indices_rf[i]], importances_rf[indices_rf[i]]))
```
